```python
import pandas as pd
from sqlalchemy import create_engine
```

----

```python
engine = create_engine('postgresql://postgres:postgres@localhost:5433/ny_taxi')
```

----

```python
df_trip = pd.read_parquet('data/green_tripdata_2025-11.parquet')

df_trip = df_trip.astype(dtype) 

for col in ['lpep_pickup_datetime', 'lpep_dropoff_datetime']:
    df_trip[col] = pd.to_datetime(df_trip[col])
```

```python
df_trip.dtypes
```

```python
print(pd.io.sql.get_schema(df_trip, name='gren_taxi_data', con=engine))
```

```python
df_trip.to_sql(name='green_taxi_data', con=engine, if_exists='append')
```

----

```python
df_zones = pd.read_csv('data/taxi_zone_lookup.csv')
```

```python
print(pd.io.sql.get_schema(df_zones, name='zones', con=engine))
```

```python
df_zones.to_sql(name='zones', con=engine, if_exists='append', index=False)
```
